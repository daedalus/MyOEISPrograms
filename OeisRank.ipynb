```python
import re
import os

directory_path = 'oeisdata/'
D = {}

def proc_file(filename, D):
  with open(filename,"r") as fp:
    for line in fp:
      line = line.strip()
      if line[:2] == "%A":
        M = re.findall(r"_([^_]+)_", line)
        for name in M:
          if name in D:
            D[name] += 1
          else:
            D[name] = 1
  return D

for root, _, files in os.walk(directory_path):
    for file in files:
        filename = os.path.join(root, file)
        if filename[-4:] == ".seq":
            D = proc_file(filename, D)
```

```python
# Sort the dictionary by value in descending order and print
L = len(D.items())
print("Percentile | rank | author | authored ")
for n, (key, value) in enumerate(sorted(D.items(), key=lambda item: item[1], reverse=True)):
    print(f'{round(((n+1)/L)*100,2)} | {n+1} | {key} | {value}')
```
